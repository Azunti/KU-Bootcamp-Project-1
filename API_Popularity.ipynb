```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
import requests
import time
from scipy.stats import linregress
from scipy import stats
import sys
from pprint import pprint
from collections import Counter
```

```python
#Billboard Dataframe 1 of 2
BB = Path('Data/BBwAPI.csv')
BB_df = pd.read_csv(BB)
BB_df = BB_df.rename(columns={"song": "Song",
                                        "performer": "Artist",
                                        "time_on_chart": "Billboard TOCM",
                                         "Popularity": "Spotify Popularity"})
BB_df["Billboard Rank"]=""
BB_df["Spotify Rank"]= BB_df.index+1
BB_df = BB_df[["Song", "Artist","Billboard TOCM","Billboard Rank", "Spotify Popularity","Spotify Rank"]]
BB_df.head()
```

```python
#Billboard Dataframe 2 of 2
BB_df = BB_df.sort_values(by='Billboard TOCM', ascending=False)
BB_df.reset_index(inplace=True)
BB_df["Billboard Rank"]= BB_df.index+1
BB_df.head()
```

```python
#Rolling Stone Dataframe
RS = Path('Data/RSwAPI.csv')
RS_df = pd.read_csv(RS)
RS_df = RS_df.rename(columns={"Rank": "Rolling Stone Rank",
                                        "Popularity": "Spotify Popularity"})
RS_df["Spotify Rank"]=RS_df.index+1
RS_df = RS_df[["Song", "Artist","Rolling Stone Rank", "Spotify Popularity","Spotify Rank"]]
RS_df.head()
```

```python
#Scatterplotting Billboard vs Spotify Rank
x_values = BB_df['Spotify Rank']
y_values = BB_df['Billboard Rank']
plt.scatter(x_values,y_values, marker='o')
x = np.linspace(0, 100, 400)
y = x
equality, = plt.plot(x,y,color='r',label='Equal Ranking')
sp_label= plt.fill_between(x, y, y2=np.max(y), where=(y <= np.max(y)), color='lightgreen', alpha=0.25, label='Higher Spotify Ranking')
bb_label = plt.fill_between(x, y,  color='m', alpha=0.1, label='Higher Billboard Ranking')
plt.title('Relationship between Billboard and Spotify Ranks')
plt.xlabel('Spotify Rank')
plt.ylabel('Billboard Rank')
plt.legend(handles=[equality, sp_label, bb_label], loc=2)
plt.savefig("output_data/BBvS.png")
plt.show()
```

```python
#Scatterplotting Rolling Stone vs Spotify Rank
x_values = RS_df['Spotify Rank']
y_values = RS_df['Rolling Stone Rank']
plt.scatter(x_values,y_values, marker='o')
x = np.linspace(0, 100, 400)
y = x
equality, = plt.plot(x,y,color='r',label='Equal Ranking')
sp_label= plt.fill_between(x, y, y2=np.max(y), where=(y <= np.max(y)),color='lightgreen', alpha=0.25, label='Higher Spotify Ranking')
rs_label = plt.fill_between(x, y,  color='bisque', alpha=0.25, label='Higher Rolling Stone Ranking')
plt.title('Relationship between Rolling Stone and Spotify Ranks')
plt.xlabel('Spotify Rank')
plt.ylabel('Rolling Stone Rank')
plt.legend(handles=[equality, sp_label, rs_label], loc=4)
plt.savefig("output_data/RSvS.png")
plt.show()
```

```python
#Experimental - Annnotated Billboard
x_values = BB_df['Spotify Rank']
y_values = BB_df['Billboard Rank']
plt.scatter(x_values,y_values, marker='o')
x = np.linspace(0, 100, 400)
y = x
equality, = plt.plot(x,y,color='r',label='Equal Ranking')
sp_label= plt.fill_between(x, y, y2=np.max(y), where=(y <= np.max(y)), color='lightgreen', alpha=0.25, label='Higher Spotify Ranking')
bb_label = plt.fill_between(x, y,  color='m', alpha=0.1, label='Higher Billboard Ranking')
accurate = 0
inaccurate = 0
for (x_val, y_val) in zip(x_values, y_values):
    # Calculate the distance from the point to the line y = x
    distance = abs(y_val - x_val)
    if distance < 1:
        plt.annotate(f'({x_val}, {y_val})',
                     (x_val, y_val),
                     textcoords="offset points",
                     xytext=(0,5),
                     ha='center')
        accurate = accurate + 1
    elif distance > 49:
         plt.annotate(f'({x_val}, {y_val})',
                     (x_val, y_val),
                     textcoords="offset points",
                     xytext=(0,5),
                     ha='center')
         inaccurate = inaccurate +1
plt.title('Relationship between Billboard and Spotify Ranks')
plt.xlabel('Spotify Rank')
plt.ylabel('Billboard Rank')
plt.legend(handles=[equality, sp_label, bb_label], loc=1)
plt.savefig("output_data/Experimental_AnnoBBvS.png")
plt.show()
print(f'There were {accurate} songs with equal ratings.')
print(f'There were {inaccurate} songs with ratings 50 or more points apart.')
```
