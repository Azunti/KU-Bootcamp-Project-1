```python
# Dependencies
import pandas as pd
from pathlib import Path
```

```python
# importing raw csv
raw_df = pd.read_csv(Path("Hot100Base.csv"))
raw_df.head()
```

```python
raw_df.shape
```

```python
#Desirable Statistics: song / performer / time_on_chart_max / peak_position
#TOCM representing overall impact, peak representing popularity
grouping_df=raw_df
del grouping_df['chart_date']
#del grouping_df['song_id']
del grouping_df['instance']
del grouping_df['consecutive_weeks']
del grouping_df['previous_week']
del grouping_df['peak_position']
del grouping_df['worst_position']
#del grouping_df['chart_debut']
del grouping_df['chart_url']
grouping_df=grouping_df.groupby(["song","performer"])
grouping_df.head()
```

```python
grouping_df.min(['chart_position'])
```

```python
grouping_df.max(['time_on_chart'])
```

```python
#group discussion: we drop peak position and we only want the top 100 of the TOCM
sorted_df=raw_df.sort_values("time_on_chart", ascending=False)
sorted_df.head()
```

```python
tocm_rank = raw_df.value_counts(['song_id'])
```

```python
tocm_rank = tocm_rank[:100]
print(tocm_rank)
```

```python
top_df = raw_df.loc[(raw_df["time_on_chart"] >= 52) , :]
top_df = top_df.sort_values("time_on_chart", ascending = False)
top_df.head(25)
```

```python
sorted_df = top_df.drop_duplicates('song_id')
del sorted_df['chart_position']
sorted_df.head(25)
```

```python
sorted_df.to_csv("tocm_rank.csv", index=False, header=True)
```
