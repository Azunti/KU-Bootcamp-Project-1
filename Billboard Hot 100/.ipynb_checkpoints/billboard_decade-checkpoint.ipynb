```python
# Dependencies
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
import requests
import time
from scipy.stats import linregress
from scipy import stats
```

```python
# importing raw csv
tocm_df = pd.read_csv(Path("tocm_rank.csv"))
tocm_df.head()
```

```python
tocm_df['chart_debut']=tocm_df['chart_debut'].str.replace('-','').apply(int)
#code for above found here: https://stackoverflow.com/questions/44929166/convert-pandas-date-object-to-integer-of-form-yyyymmdd
tocm_df.head()
```

```python
bins = [19580101, 19700101, 19800101, 19900101, 20000101, 20100101, 20200101, 20250101]
labels = ['1958-1969', '1970-1979', '1980-1989', '1990-1999','2000-2009', '2010-2019', '2020-2024']
tocm_df["Decade"]=pd.cut(tocm_df["chart_debut"],bins,labels=labels,include_lowest=True)
tocm_df.head()
```

```python
song_count = tocm_df.groupby('Decade').size().reset_index(name='Song Count')
song_count
```

```python
song_count_data = {'Decade': ['1958-1969', '1970-1979', '1980-1989', '1990-1999','2000-2009', '2010-2019','2020-2024'],
                    'Song Count': song_count['Song Count'].tolist()
                  }
decade_df = pd.DataFrame(song_count_data)
fig1 = decade_df.plot(kind='line', x='Decade', y='Song Count', marker='o', title='Amount of songs in Top 100 per Release Decade', ylabel='Total Amount of Songs',xlabel='Decade')
plt.xticks(rotation=10, rotation_mode="anchor", ha="right", wrap=True)
plt.savefig("../output_data/bbh100_decade.png")
plt.show()
```
